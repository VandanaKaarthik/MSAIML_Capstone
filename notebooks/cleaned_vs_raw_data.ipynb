```python
import os
import pandas as pd

RAW_DIR = "data/csv_files/raw_data"
CLEAN_DIR = "data/csv_files/cleaned_data"
N_TEMPLATES = 5

raw = {}
clean = {}

for i in range(1, N_TEMPLATES + 1):
    raw_path = os.path.join(RAW_DIR, f"template{i}.csv")
    clean_path = os.path.join(CLEAN_DIR, f"template{i}_clean.csv")

    raw[i] = pd.read_csv(raw_path)
    clean[i] = pd.read_csv(clean_path)

print("Loaded templates:", list(raw.keys()))
```

```python
from IPython.display import display, Markdown

N_ROWS = 5

for i in range(1, N_TEMPLATES + 1):
    display(Markdown(f"# Template {i}"))
    display(Markdown("## Raw (Top 5)"))
    display(raw[i].head(N_ROWS))

    display(Markdown("## Cleaned (Top 5)"))
    display(clean[i].head(N_ROWS))
```

```python
from collections import OrderedDict

def presence_matrix(dfs_dict):
    all_cols = sorted(set().union(*[set(df.columns) for df in dfs_dict.values()]))
    mat = pd.DataFrame(index=[f"template{i}" for i in dfs_dict.keys()], columns=all_cols)

    for i, df in dfs_dict.items():
        mat.loc[f"template{i}", :] = [1 if c in df.columns else 0 for c in all_cols]

    return mat.astype(int)

raw_presence = presence_matrix(raw)
clean_presence = presence_matrix(clean)

display(Markdown("## Raw Column Presence (1=present, 0=absent)"))
display(raw_presence)

display(Markdown("## Cleaned Column Presence (1=present, 0=absent)"))
display(clean_presence)
```

```python
def missingness_table(df, top_n=20):
    miss_pct = (df.isna().mean() * 100).sort_values(ascending=False)
    out = pd.DataFrame({"missing_%": miss_pct.round(2)})
    return out.head(top_n)

for i in range(1, N_TEMPLATES + 1):
    display(Markdown(f"# Template {i} — Missingness (Cleaned)"))
    display(missingness_table(clean[i], top_n=20))
```

```python
engineered_cols = [
    "TOTAL_AMOUNT", "SUB_TOTAL_AMOUNT", "TAX_AMOUNT", "DISCOUNT_AMOUNT",
    "CURRENCY", "DATE_NORM", "DUE_DATE_NORM"
]

rows = []
for i in range(1, N_TEMPLATES + 1):
    df = clean[i]
    row = {"template": f"template{i}"}
    for c in engineered_cols:
        row[c] = round(df[c].isna().mean() * 100, 2) if c in df.columns else 100.0
    rows.append(row)

engineered_missingness = pd.DataFrame(rows)
display(Markdown("## Missingness % for Standardized Engineered Columns (Cleaned)"))
display(engineered_missingness)
```

```python
OUT_DIR = "data/csv_files/appendix_tables"
os.makedirs(OUT_DIR, exist_ok=True)

raw_presence.to_csv(os.path.join(OUT_DIR, "raw_column_presence.csv"), index=True)
clean_presence.to_csv(os.path.join(OUT_DIR, "cleaned_column_presence.csv"), index=True)
engineered_missingness.to_csv(os.path.join(OUT_DIR, "engineered_missingness_summary.csv"), index=False)

print("Saved appendix tables to:", OUT_DIR)
```

```python
from IPython.display import display, HTML

def side_by_side_tables(df_left, df_right, left_title="Raw (Top 5)", right_title="Cleaned (Top 5)", n_rows=5):
    left_html = df_left.head(n_rows).to_html(index=False, escape=False)
    right_html = df_right.head(n_rows).to_html(index=False, escape=False)

    html = f"""
    <div style="display:flex; gap:20px; align-items:flex-start;">
      <div style="flex:1; min-width: 400px;">
        <h3 style="margin:0 0 10px 0;">{left_title}</h3>
        {left_html}
      </div>
      <div style="flex:1; min-width: 400px;">
        <h3 style="margin:0 0 10px 0;">{right_title}</h3>
        {right_html}
      </div>
    </div>
    """
    display(HTML(html))


# ---- Render for each template ----
N_TEMPLATES = 5
for i in range(1, N_TEMPLATES + 1):
    display(HTML(f"<h2 style='margin-top:25px;'>Template {i}</h2>"))
    side_by_side_tables(
        raw[i],
        clean[i],
        left_title="Raw (Top 5)",
        right_title="Cleaned (Top 5)",
        n_rows=5
    )
```

```python
from IPython.display import display, HTML

# ---- Define your "key columns" sets ----
RAW_KEY_COLS = [
    "TITLE", "DATE", "DUE_DATE", "PO_NUMBER",
    "TOTAL", "SUB_TOTAL", "TAX", "DISCOUNT",
    "GSTIN_BUYER", "GSTIN_SELLER",
    "BUYER", "SELLER_ADDRESS", "SELLER_EMAIL", "SELLER_SITE",
    "_source_file"
]

CLEAN_KEY_COLS = [
    # Engineered numeric + date fields
    "TOTAL_AMOUNT", "SUB_TOTAL_AMOUNT", "TAX_AMOUNT", "DISCOUNT_AMOUNT", "CURRENCY",
    "DATE_NORM", "DUE_DATE_NORM",

    # Buyer structured fields
    "BUYER_name", "BUYER_addr_line1", "BUYER_addr_line2", "BUYER_city_state_postal_country",
    "BUYER_tel", "BUYER_email", "BUYER_site",

    # Seller structured fields
    "SELLER_name", "SELLER_addr_line1", "SELLER_addr_line2", "SELLER_city_state_postal_country",
    "SELLER_tel", "SELLER_email", "SELLER_site",

    "_source_file"
]

def pick_existing_cols(df, wanted_cols):
    return [c for c in wanted_cols if c in df.columns]

def side_by_side_key_tables(raw_df, clean_df, n_rows=5,
                            raw_title="Raw (Key Columns, Top 5)",
                            clean_title="Cleaned (Key Columns, Top 5)"):
    raw_cols = pick_existing_cols(raw_df, RAW_KEY_COLS)
    clean_cols = pick_existing_cols(clean_df, CLEAN_KEY_COLS)

    raw_view = raw_df.loc[:, raw_cols].head(n_rows) if raw_cols else raw_df.head(n_rows)
    clean_view = clean_df.loc[:, clean_cols].head(n_rows) if clean_cols else clean_df.head(n_rows)

    # render
    raw_html = raw_view.to_html(index=False, escape=False)
    clean_html = clean_view.to_html(index=False, escape=False)

    html = f"""
    <div style="display:flex; gap:20px; align-items:flex-start;">
      <div style="flex:1; min-width: 450px;">
        <h3 style="margin:0 0 10px 0;">{raw_title}</h3>
        {raw_html}
      </div>
      <div style="flex:1; min-width: 450px;">
        <h3 style="margin:0 0 10px 0;">{clean_title}</h3>
        {clean_html}
      </div>
    </div>
    """
    display(HTML(html))


# ---- Render for each template ----
N_TEMPLATES = 5
for i in range(1, N_TEMPLATES + 1):
    display(HTML(f"<h2 style='margin-top:25px;'>Template {i} — Key Columns View</h2>"))
    side_by_side_key_tables(raw[i], clean[i], n_rows=5)
```

```python
import os
import pandas as pd

RAW_DIR = "../data/csv_files/raw_data"
CLEAN_DIR = "../data/csv_files/cleaned_data"

def show_template_columns(n_templates=5):
    for i in range(1, n_templates + 1):
        raw_path = os.path.join(RAW_DIR, f"template{i}.csv")
        clean_path = os.path.join(CLEAN_DIR, f"template{i}_clean.csv")

        print("\n" + "=" * 90)
        print(f"TEMPLATE {i}")
        print("=" * 90)

        if os.path.exists(raw_path):
            raw_df = pd.read_csv(raw_path)
            print("\nRAW COLUMNS:")
            for col in raw_df.columns:
                print(f"  - {col}")
        else:
            print("\nRAW COLUMNS: file not found")

        if os.path.exists(clean_path):
            clean_df = pd.read_csv(clean_path)
            print("\nCLEANED COLUMNS:")
            for col in clean_df.columns:
                print(f"  - {col}")
        else:
            print("\nCLEANED COLUMNS: file not found")


# Run this once
show_template_columns(n_templates=5)
```
